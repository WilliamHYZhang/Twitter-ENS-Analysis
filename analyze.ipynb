```python
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
```

Load Data

```python
with open('user-data.txt') as f:
  data = [line.strip() for line in f.readlines()]

users = []

i = 0
while i < len(data):
  user = {}
  user['category'] = data[i]
  i += 1
  user['id'] = data[i]
  i += 1
  user['name'] = data[i]
  i += 1
  user['followers_count'] = int(data[i])
  i += 1
  user['friends_count'] = int(data[i])
  i += 1
  user['valid'] = data[i] == 'VALID'
  i += 2
  users.append(user)
```

Basic Statistics

```python
def basic_statistics(valid, invalid):
  total = len(valid)+len(invalid)
  print('Total:', total)
  print('Valid:', len(valid), round(len(valid)/total*100, 2), '%')
  print('Invalid:', len(invalid), round(len(invalid)/total*100, 2), '%')
  print()

valid_users = [user for user in users if user['valid']]
invalid_users = [user for user in users if not user['valid']]

print('Population:')
basic_statistics(valid_users, invalid_users)

def filter_by_category(category, users):
  return [user for user in users if user['category'] == category]

print('Category: General')
general_valid = filter_by_category('', valid_users)
general_invalid = filter_by_category('', invalid_users)
basic_statistics(general_valid, general_invalid)

print('Category: Art')
art_valid = filter_by_category('art', valid_users)
art_invalid = filter_by_category('art', invalid_users)
basic_statistics(art_valid, art_invalid)

print('Category: NFT')
nft_valid = filter_by_category('nft', valid_users)
nft_invalid = filter_by_category('nft', invalid_users)
basic_statistics(nft_valid, nft_invalid)

print('Category: DAO')
dao_valid = filter_by_category('dao', valid_users)
dao_invalid = filter_by_category('dao', invalid_users)
basic_statistics(dao_valid, dao_invalid)
```

Follower Data

```python
def get_followers(users):
  return [user['followers_count'] for user in users]

fig, ax = plt.subplots()
plt.title('Population')
ax.boxplot([get_followers(valid_users), get_followers(invalid_users)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('General')
ax.boxplot([get_followers(general_valid), get_followers(general_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('Art')
ax.boxplot([get_followers(art_valid), get_followers(art_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('NFT')
ax.boxplot([get_followers(nft_valid), get_followers(nft_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('DAO')
ax.boxplot([get_followers(dao_valid), get_followers(dao_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()
```

Following Data

```python
def get_following(users):
  return [user['friends_count'] for user in users]

fig, ax = plt.subplots()
plt.title('Population')
ax.boxplot([get_following(valid_users), get_following(invalid_users)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('General')
ax.boxplot([get_following(general_valid), get_following(general_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('Art')
ax.boxplot([get_following(art_valid), get_following(art_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('NFT')
ax.boxplot([get_following(nft_valid), get_following(nft_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()

fig, ax = plt.subplots()
plt.title('DAO')
ax.boxplot([get_following(dao_valid), get_following(dao_invalid)], showfliers=False)
plt.xticks([1, 2], ['Valid', 'Invalid'])
plt.show()
```
